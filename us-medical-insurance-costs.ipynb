# U.S. Medical Insurance Costs

```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
def average_charge(dicts):
    charge_per_it_3={}
    charge_per_it_2={}
    for item in dicts:
        a = len(dicts[item])+1
        dicts[item]=(sum(dicts[item]))/a

    dictionary_items = dicts.items()
    sorted_items = sorted(dictionary_items)
    
    for tuples in sorted_items:
        charge_per_it_2[tuples[0]]=tuples[1]
    for item in charge_per_it_2:
        charge_per_it_3[item]=charge_per_it_2[item]
       
    return charge_per_it_3

def graph(data):
    plt.plot(data.keys(), data.values())
    plt.show()
```

```python
with open('insurance.csv') as insurance_csv:
    insurance_dict = csv.DictReader(insurance_csv)
    for row in insurance_dict:
        age.append(float(row['age']))
        sex.append(row['sex'])
        bmi.append(row['bmi'])
        smoker.append(row['smoker'])
        region.append(row['region'])
        charges.append(float(row['charges']))
```

```python
charge_per_age = {}
for n in range(len(age)):
    if age[n] in charge_per_age:
        charge_per_age[age[n]]+=[float(charges[n])]
    else:
        charge_per_age[age[n]]=[float(charges[n])]

charge_per_age=average_charge(charge_per_age)
```

```python
print(charge_per_age)
```

```python
charge_per_region = {}
for n in range(len(region)):
    if region[n] in charge_per_region:
        charge_per_region[region[n]]+=[float(charges[n])]
    else:
        charge_per_region[region[n]]=[float(charges[n])]

charge_per_region=average_charge(charge_per_region)
print(charge_per_region)
```

```python
charge_per_bmi = {}
for n in range(len(bmi)):
    if bmi[n] in charge_per_bmi:
        charge_per_bmi[bmi[n]]+=[float(charges[n])]
    else:
        charge_per_bmi[bmi[n]]=[float(charges[n])]

charge_per_bmi=average_charge(charge_per_bmi)
print(charge_per_bmi)
```

```python
charge_per_sex = {}
for n in range(len(sex)):
    if sex[n] in charge_per_sex:
        charge_per_sex[sex[n]]+=[float(charges[n])]
    else:
        charge_per_sex[sex[n]]=[float(charges[n])]

charge_per_sex=average_charge(charge_per_sex)
print(charge_per_sex)
```
